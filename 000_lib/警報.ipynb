# 台風の暴風警報の解除監視プログラム

1分毎に気象庁のサイトにアクセスして、警報が消えたら、beep音を鳴らす  

```python
import pandas as pd
import datetime,time
import winsound
while True:
    url = 'http://www.jma.go.jp/jp/warn/329_table.html'
    fetched_dataframes = pd.io.html.read_html(url)
    df=fetched_dataframes[6]
    print(datetime.datetime.now(),df.iat[6,3])
    time.sleep(60)
    if df.iat[6,3]!='●':
        winsound.Beep(1000,800)
        break
```
